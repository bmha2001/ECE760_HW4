```python
import os
import math
from collections import defaultdict
```

```python
data_path = "languageID"  # Adjust the path to the dataset folder
training_files = {"e": [], "j": [], "s": []}
vocabulary = set("abcdefghijklmnopqrstuvwxyz ")  # 27 character types

for root, dirs, files in os.walk(data_path):
    for filename in files:
        lang = filename[0]
        num = int(filename.replace("e", "").replace(".txt", "").replace("j", "").replace("s", ""))
        if lang in training_files and num < 10:
            with open(os.path.join(root, filename), 'r', encoding='utf-8') as file:
                text = file.read()
                training_files[lang].append(text)
vocabulary = set("abcdefghijklmnopqrstuvwxyz ")  # 27 character types
char_probs = {lang: {} for lang in ['e', 'j', 's']}
total_chars =  {"e": 0, "j": 0, "s": 0}
char_counts = {"e": {char: 0 for char in vocabulary}, "j": {char: 0 for char in vocabulary}, "s": {char: 0 for char in vocabulary}}  # Start with additive smoothing
```

```python
for lang in training_files:
    for file in training_files[lang]:
        for char in file:
            if char in vocabulary:
                char_counts[lang][char] += 1
                total_chars[lang] += 1
```

```python
# Step 2: Estimate prior probabilities with additive smoothing (parameter 1/2)
num_classes = len(training_files)
total_training_docs = sum(len(training_files[lang]) for lang in training_files)
prior_probabilities = {lang: (len(training_files[lang]) + 0.5) / (total_training_docs + num_classes * 0.5) for lang in training_files}
for lang, prior_prob in prior_probabilities.items():
    print(f'P(y={lang}) = {prior_prob:.4f}')
```

```python
conditional_probabilities = defaultdict(lambda: defaultdict(float))
for lang in training_files:
    char_counts = defaultdict(int)
    for doc in training_files[lang]:
        for char in doc:
            if char in vocabulary:
                char_counts[char] += 1
    total_chars = sum(char_counts.values())
    for char in vocabulary:
        conditional_probabilities[lang][char] = (char_counts[char] + 0.5) / (total_chars + len(vocabulary) * 0.5)
```

```python
# Load and preprocess the test document "e10.txt"
test_document_path = "languageID/e10.txt"  # Adjust the path to the test document
with open(test_document_path, 'r', encoding='utf-8') as file:
    test_document = file.read()

# Initialize the bag-of-words count vector
vocabulary = "abcdefghijklmnopqrstuvwxyz "  # 27 character types
bag_of_characters = {char: 0 for char in vocabulary}

# Count the occurrences of characters in the test document
for char in test_document:
    if char in bag_of_characters:
        bag_of_characters[char] += 1

# Print the bag-of-words vector x
print("Bag-of-Characters Vector x:")
for char, count in bag_of_characters.items():
    print(f"{char}: {count}", end=", ")
```

```python
# Define the bag-of-words (bag-of-characters) vector for the test document x
# This vector should be based on your "e10.txt" test document
test_document_vector = bag_of_characters.values()
likelihood_probs = {"e": 0, "j": 0, "s": 0}

# Calculate p(x | y) for y = e
prob_x_given_e = 1
for char, count in zip(vocabulary, test_document_vector):
    prob_x_given_e += (conditional_probabilities['e'][char] * count)
    likelihood_probs['e'] = prob_x_given_e

# Calculate p(x | y) for y = j
prob_x_given_j = 1
for char, count in zip(vocabulary, test_document_vector):
    prob_x_given_j += (conditional_probabilities['j'][char] * count)
    likelihood_probs['j'] = prob_x_given_j
    

# Calculate p(x | y) for y = s
prob_x_given_s = 1
for char, count in zip(vocabulary, test_document_vector):
    prob_x_given_s += (conditional_probabilities['s'][char] * count)
    likelihood_probs['s'] = prob_x_given_s

# Print the probabilities
print("ˆp(x | y = e):", prob_x_given_e)
print("ˆp(x | y = j):", prob_x_given_j)
print("ˆp(x | y = s):", prob_x_given_s)
```

```python
posterior_probs = {lang: likelihood_probs[lang] * 0.3 for lang in ['e', 'j', 's']}

# Normalize posterior probabilities to sum to 1
total_posterior_prob = sum(posterior_probs.values())
normalized_posterior_probs = {lang: prob / total_posterior_prob for lang, prob in posterior_probs.items()}

# Predict the class label
predicted_class = max(normalized_posterior_probs, key=normalized_posterior_probs.get)

# Print posterior probabilities and predicted class label
for lang in ['e', 'j', 's']:
    print(f'ˆp(y = {lang} | x) = {normalized_posterior_probs[lang]:.4f}')
print(f'Predicted Class Label for e10.txt: {predicted_class}')
```
